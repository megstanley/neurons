# Inspecting time series data

This is an initial inspection of the time series data extracted from 2 Hz full brain sections of the larval zebrafish, 
recorded for a total of 1815 frames, or approximately 15 minutes. First, let's understand the structure of the data! 

The ultimate aim is to try to identify circuits of neurons that are associated in resting state behaviour, and be able to draw a connectivity graph. 

First, let's import some useful packages.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = np.load('/Users/Megan/data_analysis/Aug02_2018_B3/B3_TS/Extracted/rg_B3_TS_ZP_1.npz')
```

First, understand the arrangement of the data. An npz file suggests we have data stored in arrays, but it is in fact a dictionary of arrays, to which we do not yet know the keys.

```python
data.keys()
```

```python
n_data = data['data']
n_data.shape
```

```python
xy_data = data['xy']
xy_data.shape
```

```python
xy_data[23,:]
```

We know there are 1815 time points in each of the movies, and the other axis represents the separate neurons. We first do some inspectional plotting of a time series. The xy (and z) data will be dealt with below.

```python
timepoints = range(0, 1815)
timepoints = np.divide(timepoints, 2.0) #account for 2Hz sampling rate.
```

```python
plt.plot(timepoints[:], n_data[:,1])
plt.xlabel('Time (s)')
plt.ylabel('Signal')
```

So we have very messy single neuron firing data for a single z-slice of the brain, over time. Let's just take a look at a few other neurons to see if we can understand more generally what the data is like. 

```python
fig, ax = plt.subplots(10, sharex='col', figsize = (10,5))
fig.suptitle('Full time series for ten neurons in the same layer')
for i in range(20, 30):
    ax[i-20].plot(timepoints, n_data[:,i]) 
fig.subplots_adjust(hspace=0)

for axis in ax:
    axis.label_outer()
```

Actually, a heatmap would be nice here to see all the neurons in our data set at once, and see if we can spot any initial trends or missing data. 

```python
fig, heat_map = plt.subplots(figsize=(10,5))
sns.heatmap(n_data, ax = heat_map, yticklabels = 100)
heat_map.set_title('All neurons in z slice')
heat_map.set_xlabel('neuron')
heat_map.set_ylabel('Time')
```

Several of the neurons seem to go completely dark (seem to stop firing at all) after approximately the first 600 seconds. It is not clear whether this is an experimental problem (photobleaching, or drifting out of the frame) or if it is a true effect. We can simplify matters in the below by ignoring time points after 1200.

We are first going to look for any directly time-correlated behaviour between the neurons; that is, do they exhibit correlations at the same time (indicating that they are closely connected in a graph model), and do they belong to any groups? However, the data are not yet clean enough.

## Baseline subtraction

We can first try subtracting a baseline, which is just the smallest value of the time series array for each neuron. This is a very simple first approach, but it might be a good way to improve the correlations. At this point we are just trying things out to understand the data. 

```python
n_data = n_data[0:1200, :] #remove the data where many neurons are known to show no signal at all.
```

```python
n_data.shape
```

```python
for i in range(307):
   mini = min(n_data[:,i])
   n_data[:,i] = n_data[:,i] - mini
```

```python
fig, heat_map = plt.subplots(figsize=(10,5))
sns.heatmap(n_data, ax = heat_map, yticklabels = 100)
heat_map.set_title('All neurons in z slice, baseline subtracted')
heat_map.set_xlabel('neuron')
heat_map.set_ylabel('Time')
```

```python
fig, ax = plt.subplots(20, sharex='col', figsize = (20,10))
fig.suptitle('Full time series for ten neurons in the same layer, baseline subtracted')
fig.titlesize = 40
for i in range(20):
    ax[i].plot(timepoints[0:1200], n_data[:,i]) 
fig.subplots_adjust(hspace=0)

for axis in ax:
    axis.label_outer()
```

This data has a combination of neurons that appear to just have a background signal all the time, and those which have some activity, however noisy, that may be correlated to the others. 

It might be possible to analyse this data by looking for the derivatives, as that will act as a basic filer to remove DC components. So let's have a look at that, where we normalise by the magnitude of the signal to perform crude correction (for local brightness due to alignment of microscope/sensitivity of camera pixels etc)

```python
derivs = np.gradient(n_data[:,1])
normedderivs = np.divide(derivs, n_data[:,1])
plt.plot(timepoints[0:1200], normedderivs)
```

```python
derivs = np.gradient(n_data[:,5])
plt.plot(timepoints[0:1200], derivs)
```

```python
fig, ax = plt.subplots(20, sharex='col', figsize = (20,10))
fig.suptitle('Full time series for ten neurons in the same layer, baseline subtracted')
fig.titlesize = 40
for i in range(20):
    derivs = np.gradient(n_data[:,i])
    normedderiv = np.divide(derivs, n_data[:,i])
    ax[i].plot(timepoints[0:1200], normedderiv) 
fig.subplots_adjust(hspace=0)

for axis in ax:
    axis.label_outer()
```

Hmmm. This approach may be a bit crude.
